Calculate the mean, median, mode for petal length

```python
# mean for petal length:
# mean_petal_length = df['petal_length'].mean().round(2)
mean_petal_length = df['petal_length'].mean()
print("Mean petal length:", mean_petal_length) 
# median for petal width:
median_petal_width = df['petal_width'].median()
print("Median petal width:", median_petal_width)
# mode for sepal length:
mode_petal_length = df['petal_length'].mode()[0]
print("Mode petal length:", mode_petal_length)
```

What can you conclude about the distribution for petal length?

With a mean petal length of 3.76 and a mode of 1.5, the distribution doesn’t seem symmetric. This indicates that petal length is spread across distinct groups (showing peaks depending on species), so the distribution will not follow a simple bell curve. So instead of one smooth bell curve, you should get separate peaks.

What is the smallest and largest value for petal length?

```python
# use min() and max() on the column
smallest = df["petal_length"].min()
largest = df["petal_length"].max()

print("Smallest petal length:", smallest)
print("Largest petal length:", largest)
```

Calculate the variance and standard deviation for petal length

```python
# variance and standard deviation
variance_petal_length = df["petal_length"].var()
stddev_petal_length = df["petal_length"].std()
print("Variance of petal length:", variance_petal_length)
print("Standard deviation of petal length:", stddev_petal_length)
```

Create a new column with a rough estimate of petal area by multiplying petal length and width together.

Create a new DataFrame with petal areas greater than $1cm^2$.

Create 3 new DataFrames, one for each species. Use the entire dataset.

```python
# Normalize species names
df["species_norm"] = (
    df["species"]
    .str.strip()
    .str.lower()
    .str.replace(r"\s+", "-", regex=True)  # spaces -> hyphen
)

# Optional: map to canonical names
name_map = {
    "iris-setosa": "Iris-setosa",
    "iris-versicolor": "Iris-versicolor",
    "iris-virginica": "Iris-virginica",
    "setosa": "Iris-setosa",
    "versicolor": "Iris-versicolor",
    "virginica": "Iris-virginica",
}
df["species_canonical"] = df["species_norm"].map(name_map).fillna(df["species"])

# Split using canonical labels
df_setosa = df[df["species_canonical"] == "Iris-setosa"]
df_versicolor = df[df["species_canonical"] == "Iris-versicolor"]
df_virginica = df[df["species_canonical"] == "Iris-virginica"]

# Verify counts
print(df["species_canonical"].value_counts())

df_setosa = df[df["species"] == "Iris-setosa"]
df_versicolor = df[df["species"] == "Iris-versicolor"]
df_virginica = df[df["species"] == "Iris-virginica"]

print("Setosa:\n", df_setosa.head(), "\n")
print("Versicolor:\n", df_versicolor.head(), "\n")
print("Virginica:\n", df_virginica.head())




```
